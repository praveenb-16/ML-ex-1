```python
import csv
import numpy as np
import pandas as pd

# Step 1: Create CSV file automatically (so no FileNotFoundError)
with open("finds1.csv", "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["Sky", "AirTemp", "Humidity", "Wind", "Water", "Forecast", "EnjoySport"])
    writer.writerow(["Sunny", "Warm", "Normal", "Strong", "Warm", "Same", "Yes"])
    writer.writerow(["Sunny", "Warm", "High", "Strong", "Warm", "Same", "Yes"])
    writer.writerow(["Rainy", "Cold", "High", "Strong", "Warm", "Change", "No"])
    writer.writerow(["Sunny", "Warm", "High", "Strong", "Cool", "Change", "Yes"])

# Step 2: Read dataset
data = pd.read_csv("finds1.csv")

# Step 3: Separate features and target
concepts = np.array(data.iloc[:, 0:-1])
target = np.array(data.iloc[:, -1])

# Step 4: Candidate Elimination Algorithm
def learn(concepts, target):

    # Initialize specific hypothesis
    specific_h = concepts[0].copy()
    print("Initialization of specific_h:")
    print(specific_h)

    # Initialize general hypothesis
    general_h = [["?" for _ in range(len(specific_h))] for _ in range(len(specific_h))]
    print("\nInitialization of general_h:")
    for row in general_h:
        print(row)

    # Process each training example
    for i, h in enumerate(concepts):
        print("\nStep", i + 1)

        if target[i] == "Yes":  # Positive example
            for x in range(len(specific_h)):
                if h[x] != specific_h[x]:
                    specific_h[x] = "?"
                    general_h[x][x] = "?"

        elif target[i] == "No":  # Negative example
            for x in range(len(specific_h)):
                if h[x] != specific_h[x]:
                    general_h[x][x] = specific_h[x]
                else:
                    general_h[x][x] = "?"

        # Print step results
        print("Specific_h:", specific_h)
        print("General_h:")
        for row in general_h:
            print(row)

    # Remove rows with all '?'
    general_h = [row for row in general_h if row != ["?", "?", "?", "?", "?", "?"]]

    return specific_h, general_h


# Step 5: Run algorithm
s_final, g_final = learn(concepts, target)

# Step 6: Final Output
print("\nFinal Specific Hypothesis:")
print(s_final)

print("\nFinal General Hypothesis:")
for row in g_final:
    print(row)
```
