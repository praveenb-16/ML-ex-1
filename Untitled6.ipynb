```python
import numpy as np
import pandas as pd
import csv

from pgmpy.models import DiscreteBayesianNetwork
from pgmpy.estimators import MaximumLikelihoodEstimator
from pgmpy.inference import VariableElimination


# -----------------------------
# Step 1: Create heart.csv automatically
# -----------------------------
with open("heart.csv", "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow([63, 1, 1, 145, 233, 1, 2, 150, 0, 2.3, 3, 0, 6, 1])
    writer.writerow([37, 1, 2, 130, 250, 0, 1, 187, 0, 3.5, 2, 0, 3, 0])
    writer.writerow([41, 0, 1, 130, 204, 0, 2, 172, 0, 1.4, 1, 0, 3, 0])
    writer.writerow([56, 1, 1, 120, 236, 0, 1, 178, 0, 0.8, 1, 0, 3, 1])
    writer.writerow([57, 0, 0, 120, 354, 0, 1, 163, 1, 0.6, 1, 0, 3, 1])
    writer.writerow([57, 1, 0, 140, 192, 0, 1, 148, 0, 0.4, 2, 0, 6, 0])
    writer.writerow([56, 0, 1, 140, 294, 0, 2, 153, 0, 1.3, 2, 0, 3, 0])
    writer.writerow([44, 1, 1, 120, 263, 0, 1, 173, 0, 0.0, 1, 0, 7, 0])


# -----------------------------
# Step 2: Define column names
# -----------------------------
names = [
    'age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg',
    'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal', 'heartdisease'
]

# -----------------------------
# Step 3: Load dataset
# -----------------------------
heartDisease = pd.read_csv("heart.csv", names=names)

# Replace missing values if any
heartDisease = heartDisease.replace("?", np.nan)

print("Dataset loaded successfully!\n")
print(heartDisease.head())


# -----------------------------
# Step 4: Build Bayesian Model
# -----------------------------
model = DiscreteBayesianNetwork([
    ('age', 'trestbps'),
    ('age', 'fbs'),
    ('sex', 'trestbps'),
    ('exang', 'trestbps'),
    ('trestbps', 'heartdisease'),
    ('fbs', 'heartdisease'),
    ('heartdisease', 'restecg'),
    ('heartdisease', 'thalach'),
    ('heartdisease', 'chol')
])

# -----------------------------
# Step 5: Train the model
# -----------------------------
model.fit(heartDisease, estimator=MaximumLikelihoodEstimator)

print("\nModel trained successfully!")

# -----------------------------
# Step 6: Inference
# -----------------------------
HeartDisease_infer = VariableElimination(model)

q = HeartDisease_infer.query(
    variables=['heartdisease'],
    evidence={'age': 37, 'sex': 1}
)

print("\nProbability of Heart Disease:")
print(q)
```
