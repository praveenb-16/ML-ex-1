```python
import numpy as np
import math
import csv


# -----------------------------
# Read data from file
# -----------------------------
def read_data(filename):
    with open(filename, 'r') as file:
        reader = csv.reader(file)
        data = list(reader)

    metadata = np.array(data[0])       # Header row
    traindata = np.array(data[1:])     # Actual data rows
    return metadata, traindata


# -----------------------------
# Node class for Decision Tree
# -----------------------------
class Node:
    def __init__(self, attribute):
        self.attribute = attribute
        self.children = []
        self.answer = ""

    def __str__(self):
        return self.attribute


# -----------------------------
# Create subtables
# -----------------------------
def subtables(data, col, delete):
    dic = {}
    items = np.unique(data[:, col])
    count = np.zeros((items.shape[0], 1), dtype=np.int32)

    for x in range(items.shape[0]):
        for y in range(data.shape[0]):
            if data[y, col] == items[x]:
                count[x] += 1

    for x in range(items.shape[0]):
        dic[items[x]] = np.empty((int(count[x]), data.shape[1]), dtype=object)
        pos = 0
        for y in range(data.shape[0]):
            if data[y, col] == items[x]:
                dic[items[x]][pos] = data[y]
                pos += 1
        if delete:
            dic[items[x]] = np.delete(dic[items[x]], col, 1)

    return items, dic


# -----------------------------
# Entropy Calculation
# -----------------------------
def entropy(S):
    items = np.unique(S)

    if items.size == 1:
        return 0

    counts = np.zeros((items.shape[0], 1))
    sums = 0

    for x in range(items.shape[0]):
        counts[x] = sum(S == items[x]) / (S.size * 1.0)

    for count in counts:
        sums += -1 * count * math.log(count, 2)

    return sums


# -----------------------------
# Gain Ratio Calculation
# -----------------------------
def gain_ratio(data, col):
    items, dic = subtables(data, col, delete=False)
    total_size = data.shape[0]

    entropies = np.zeros((items.shape[0], 1))
    intrinsic = np.zeros((items.shape[0], 1))

    for x in range(items.shape[0]):
        ratio = dic[items[x]].shape[0] / (total_size * 1.0)
        entropies[x] = ratio * entropy(dic[items[x]][:, -1])
        intrinsic[x] = ratio * math.log(ratio, 2)

    total_entropy = entropy(data[:, -1])
    iv = -1 * sum(intrinsic)

    for x in range(entropies.shape[0]):
        total_entropy -= entropies[x]

    if iv == 0:
        return 0

    return total_entropy / iv


# -----------------------------
# Create Decision Tree
# -----------------------------
def create_node(data, metadata):
    # If all outputs are same, return leaf node
    if (np.unique(data[:, -1])).shape[0] == 1:
        node = Node("")
        node.answer = np.unique(data[:, -1])[0]
        return node

    gains = np.zeros((data.shape[1] - 1, 1))

    for col in range(data.shape[1] - 1):
        gains[col] = gain_ratio(data, col)

    split = np.argmax(gains)

    node = Node(metadata[split])
    new_metadata = np.delete(metadata, split, 0)

    items, dic = subtables(data, split, delete=True)

    for x in range(items.shape[0]):
        child = create_node(dic[items[x]], new_metadata)
        node.children.append((items[x], child))

    return node


# -----------------------------
# Print spaces
# -----------------------------
def empty(size):
    s = ""
    for x in range(size):
        s += " "
    return s


# -----------------------------
# Print Decision Tree
# -----------------------------
def print_tree(node, level):
    if node.answer != "":
        print(empty(level), node.answer)
        return

    print(empty(level), node.attribute)

    for value, n in node.children:
        print(empty(level + 1), value)
        print_tree(n, level + 2)


# -----------------------------
# Create tennis.data automatically
# -----------------------------
with open("tennis.data", "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["Outlook", "Temperature", "Humidity", "Wind", "PlayTennis"])
    writer.writerow(["Sunny", "Hot", "High", "Weak", "No"])
    writer.writerow(["Sunny", "Hot", "High", "Strong", "No"])
    writer.writerow(["Overcast", "Hot", "High", "Weak", "Yes"])
    writer.writerow(["Rain", "Mild", "High", "Weak", "Yes"])
    writer.writerow(["Rain", "Cool", "Normal", "Weak", "Yes"])
    writer.writerow(["Rain", "Cool", "Normal", "Strong", "No"])
    writer.writerow(["Overcast", "Cool", "Normal", "Strong", "Yes"])
    writer.writerow(["Sunny", "Mild", "High", "Weak", "No"])
    writer.writerow(["Sunny", "Cool", "Normal", "Weak", "Yes"])
    writer.writerow(["Rain", "Mild", "Normal", "Weak", "Yes"])
    writer.writerow(["Sunny", "Mild", "Normal", "Strong", "Yes"])
    writer.writerow(["Overcast", "Mild", "High", "Strong", "Yes"])
    writer.writerow(["Overcast", "Hot", "Normal", "Weak", "Yes"])
    writer.writerow(["Rain", "Mild", "High", "Strong", "No"])


# -----------------------------
# Main Program
# -----------------------------
metadata, traindata = read_data("tennis.data")
data = np.array(traindata, dtype=object)

node = create_node(data, metadata[:-1])   # exclude class label
print("Decision Tree:\n")
print_tree(node, 0)
```
