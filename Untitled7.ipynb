```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
import csv


# -----------------------------
# Step 1: Create kmeansdata.csv automatically
# -----------------------------
with open("kmeansdata.csv", "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["Distance_Feature", "Speeding_Feature"])
    writer.writerow([10, 5])
    writer.writerow([12, 6])
    writer.writerow([15, 8])
    writer.writerow([50, 30])
    writer.writerow([55, 28])
    writer.writerow([58, 32])
    writer.writerow([80, 45])
    writer.writerow([85, 42])
    writer.writerow([90, 48])


# -----------------------------
# Step 2: Read dataset
# -----------------------------
X = pd.read_csv("kmeansdata.csv")

x1 = X['Distance_Feature'].values
x2 = X['Speeding_Feature'].values

X = np.array(list(zip(x1, x2))).reshape(len(x1), 2)


# -----------------------------
# Step 3: Plot original dataset
# -----------------------------
plt.figure(figsize=(6, 5))
plt.xlim([0, 100])
plt.ylim([0, 50])
plt.title('Dataset')
plt.scatter(x1, x2)
plt.xlabel("Distance Feature")
plt.ylabel("Speeding Feature")
plt.show()


# -----------------------------
# Step 4: EM Algorithm using Gaussian Mixture Model
# -----------------------------
gmm = GaussianMixture(n_components=3, random_state=42)
gmm.fit(X)

em_predictions = gmm.predict(X)

print("\nEM Predictions")
print(em_predictions)

print("\nMeans:")
print(gmm.means_)

print("\nCovariances:")
print(gmm.covariances_)

print("\nDataset:")
print(X)

plt.figure(figsize=(6, 5))
plt.title('Expectation Maximization (EM)')
plt.scatter(X[:, 0], X[:, 1], c=em_predictions, s=50, cmap='viridis')
plt.xlabel("Distance Feature")
plt.ylabel("Speeding Feature")
plt.show()


# -----------------------------
# Step 5: K-Means Clustering
# -----------------------------
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)

print("\nK-Means Cluster Centers:")
print(kmeans.cluster_centers_)

print("\nK-Means Labels:")
print(kmeans.labels_)

plt.figure(figsize=(6, 5))
plt.title('K-Means Clustering')
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, cmap='rainbow', s=50)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], color='black', s=200, marker='X')
plt.xlabel("Distance Feature")
plt.ylabel("Speeding Feature")
plt.show()
```
