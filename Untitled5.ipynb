```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn import metrics
import csv


# Step 1: Create CSV file automatically
with open("naivetext1.csv", "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["I love this movie", "pos"])
    writer.writerow(["This film was amazing", "pos"])
    writer.writerow(["I really liked the acting", "pos"])
    writer.writerow(["What a fantastic story", "pos"])
    writer.writerow(["I hate this movie", "neg"])
    writer.writerow(["This film was terrible", "neg"])
    writer.writerow(["Worst acting ever", "neg"])
    writer.writerow(["I did not like the story", "neg"])


# Step 2: Read dataset
msg = pd.read_csv('naivetext1.csv', names=['message', 'label'])

print("The dimensions of the dataset:", msg.shape)

# Step 3: Convert labels into numbers
msg['labelnum'] = msg.label.map({'pos': 1, 'neg': 0})

# Step 4: Define input and output
X = msg.message
y = msg.labelnum

print("\nMessages:")
print(X)

print("\nLabels:")
print(y)

# Step 5: Split dataset
xtrain, xtest, ytrain, ytest = train_test_split(X, y, test_size=0.25, random_state=1)

print("\nTraining and Testing Shapes:")
print("xtest shape:", xtest.shape)
print("xtrain shape:", xtrain.shape)
print("ytest shape:", ytest.shape)
print("ytrain shape:", ytrain.shape)

# Step 6: Convert text into numerical form
count_vect = CountVectorizer()
xtrain_dtm = count_vect.fit_transform(xtrain)
xtest_dtm = count_vect.transform(xtest)

# Step 7: Train Naive Bayes model
clf = MultinomialNB()
clf.fit(xtrain_dtm, ytrain)

# Step 8: Predict test data
predicted = clf.predict(xtest_dtm)

# Step 9: Evaluation
print("\nAccuracy Metrics")
print("Accuracy of the classifier is:", metrics.accuracy_score(ytest, predicted))

print("\nConfusion Matrix")
print(metrics.confusion_matrix(ytest, predicted))

print("\nRecall and Precision")
print("Recall:", metrics.recall_score(ytest, predicted))
print("Precision:", metrics.precision_score(ytest, predicted))
```
